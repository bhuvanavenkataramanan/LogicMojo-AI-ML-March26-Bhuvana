## 2. Text Cleaning Pipeline

You are given:

```python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
```

Task:

1. Write a function clean_text(text):
2. Remove special characters
3. Remove extra spaces
4. Convert text to lowercase
5. Apply the function to all elements using loop or map
6. Return cleaned list

---

````python
import re

def clean_text(text):
    result = []
    for item in text:
        item = item.strip()
        clean_text = re.findall("[a-zA-Z]+",item)
        s = " ".join(clean_text)
        result.append(s.lower())
        
    return result
    

texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]

print("Cleaned Text: ",clean_text(texts))
````

## 7. Dictionary-Based Recommendation System

You are given:

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
```

Task:

1. Write a function recommend(user)
2. Recommend items based on common items between users
3. Do not recommend items already purchased by the user
4. Use sets and dictionary iteration

---

````python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}

def recommend(user):
    user_purchase = set(user.get(next(iter(user))))
    recommend_purchase = set()
    for k in user_purchases:
        if k == next(iter(user)):
            continue
        else:
            purchase = set(user_purchases.get(k))
            if user_purchase & purchase:
                recommend_purchase |= purchase.difference(user_purchase)

    return list(recommend_purchase)
    
for k,v in user_purchases.items():
    print(f"Recommended purchase for {k}", "is: ", recommend({k:v}))
````

## 8. Recursive Decision Tree Simulation

Task:

1. Write a recursive function decision_tree(score)
2. If score is greater than 80 return "Approved"
3. If score is between 50 and 80 call the function again with score plus 10
4. If score is less than 50 return "Rejected"

---


````python
def decision_tree(score):
    if score > 80:
        return "Approved"
    elif score >=50 and score <=80:
        return decision_tree(score+10)
    else:
        return "Rejected"
        
try:
    score = int(input("Enter a score: "))
    print(score)
    result = decision_tree(score)
    print(result)
except ValueError:
    print("Enter a valid number")
````
