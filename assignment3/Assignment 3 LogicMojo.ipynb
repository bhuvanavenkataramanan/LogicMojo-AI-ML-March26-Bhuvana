Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
from sklearn.datasets import load_iris
import numpy as np

# Load dataset
iris = load_iris()

#iris.feature_names

dataset = iris.data

#print(dataset.shape)

mean = np.mean(dataset, axis = 0)
median = np.median(dataset, axis = 0)
std = np.std(dataset, axis = 0)
var = np.var(dataset, axis = 0)

print("\nMean : ", mean)
print("Median : ", median)
print("Standard Deviation : ", std)
print("Variance : ", var)
      
max_var_idx = np.argmax(var)
#print(max_var_idx)
feature_names = iris.feature_names
print("Feature with highest variance: ",feature_names[max_var_idx])
print("Highest variance value: ",var[max_var_idx])

feature = dataset[:,0]
print("Original Shape: ",feature.shape)

feature_reshaped = feature.reshape(-1,1)
print("Original ReShape: ",feature_reshaped.shape)

print("Variance means that there will be differentiation in data for learning")
print("ML Models always expects data to be a table(2D input): Sample and Feature")
```

Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain:
Why different strategies are used for different columns

```python

import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")

#Task 1
age_median = df["age"].median()
df["age_updated"] = df["age"].fillna(age_median)
print(df["age_updated"])

#Task 2
embarked_mode = df["embarked"].mode()
df["embarked_updated"] = df["embarked"].fillna(embarked_mode)
print(df["embarked_updated"])

#Task 3
df.drop(columns = ["deck"],inplace = True)
```

Task 4

Why different strategies are used for different columns?
1. For missing age median gives a more stable (central or middle) value than mean
2. For embarked we used mode because it gives the most frequently occuring port, and making a assumption that most passengers would embark in that port
3. Of the total 891 rows, the missing value for deck is 688, which is ~77% of the rows. Hence we choose to drop it

Q11. End-to-End ML Preprocessing Pipeline (Titanic)

Context: Prepare dataset for ML model training

Tasks:

1. Load dataset
2. Handle missing values
3. Encode categorical variables (sex, embarked)
4. Normalize numerical features
5. Convert to NumPy array

Output:

Final feature matrix ready for model training

```python
#loads data
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset("titanic")

#handles missing values
df["age"] = df["age"].fillna(df["age"].median())
df["embarked"] = df["embarked"].fillna(df["embarked"].mode())

#encodes categoricals (sex, embarked)
df["sex"] = df["sex"].map({"male": 0, "female": 1})

#one-hot 
df = pd.get_dummies(df, columns=["embarked"], drop_first=True)

#normalizes numerics
#Using StandardScaler from sklearn. Another method is to compute for each feature like it was done in Question 3.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ["age", "fare", "sibsp", "parch", "pclass"]

df[num_cols] = scaler.fit_transform(df[num_cols])
print(df.head())

#outputs a NumPy array

features = df.drop("survived", axis=1)
target = df["survived"]

features_array = features.to_numpy()
target_array = target.to_numpy()

print(features_array.shape,target_array.shape)
features_array
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")

#Create age_group: Child / Adult / Senior
#Use .apply()

def age_group(age):
    if pd.isnull(age):
        return "Unknown"
    elif age < 16:
        return "Child"
    elif age <= 60:
        return "Adult"
    else:
        return "Senior"

df["age_group"] = df["age"].apply(age_group)

#Compute survival rate per group
survival_by_age_group = df.groupby("age_group")["survived"].mean()
print(survival_by_age_group)

#Interpret: Which segment has highest survival likelihood
print(survival_by_age_group.idxmax(),survival_by_age_group.max())
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

#Generate 1000 random error values using NumPy
rng = np.random.default_rng()
errors = rng.normal(loc=0.0, scale=1.0, size=1000)

#Compute: Mean Standard deviation
mean = np.mean(errors)
std = np.std(errors)

print("Mean: ", mean)
print("Standard Deviation: ", std)

#Identify outliers: Using mean ± 2*std
low = mean - 2 * std
high = mean + 2 * std

outliers = errors[(errors < low) | (errors > high)]

print("Number of Outliers:", len(outliers))
print("Outliers:", outliers)

#Explain: What high variance means in model performance
print("It is difficult to generalize high variance data")
```
